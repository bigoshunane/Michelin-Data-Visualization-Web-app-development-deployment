```python
from forex_python.converter import CurrencyRates
import pandas as pd
import datetime
```

```python
# import csv

michelin = '../data/2021CountryContinent.csv'
michelin_df = pd.read_csv(michelin, encoding='ISO-8859-1')

michelin_df.head()
```

```python
michelin_df['date'] = '2021-01-01'
michelin_df['date'] = michelin_df['date'].astype('datetime64[ns]')
```

```python
michelin_df.head()
```

```python
c = CurrencyRates()

michelin_df['Max_USD'] = michelin_df.apply( lambda x: c.convert( x.Currency, 'USD', x.MaxPrice, x.date), axis = 1)

michelin_df.head()
```

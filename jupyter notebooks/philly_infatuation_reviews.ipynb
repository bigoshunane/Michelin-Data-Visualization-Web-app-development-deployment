```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from flask_pymongo import PyMongo
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import pprint
```

```python
infatuation_url = 'https://www.theinfatuation.com/'

infatuation_review_urls = pd.read_csv('../data/philly_infatuation_review_urls.csv')  

infatuation_review_urls.drop_duplicates(subset=['urls'])
```

```python
#set up browser and parser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# set up collection
count = 0
infatuation_reviews = []

for x in infatuation_review_urls['urls']:
    browser.visit(infatuation_url + x)
    time.sleep(1)
    html = browser.html
    soup = bs(html, 'html.parser')
    try:
        infatuation_reviews.append([soup.find_all('div', class_="styles_richText__a6C_c")])
    except:
        infatuation_reviews.append('Null')
        print('Null')
    count = count + 1
    print(count)
    
print('Finished')
```

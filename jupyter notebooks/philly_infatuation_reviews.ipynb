```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from flask_pymongo import PyMongo
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import pprint


#scraping urls

philly_rest_url = 'https://www.phillymag.com/foobooz/50-best-restaurants/'


#List to store text for restaurant reviews.
reviews=[] 

#set up browser and parser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

browser.visit(philly_rest_url)

html = browser.html
soup = bs(html, 'html.parser')
   

# set up collection

#for a in soup.findAll('a',href=True, attrs={'class':'_31qSD5'}):
for a in soup.findAll('div', attrs={'class': 'wysiwyg'}):
    
    # getting all the paragraphs
    for para in soup.find_all("p"):
        #print(para.get_text())

        reviews.append(para.get_text())

print('Finished')
```

```python
#scraping urls

infatuation_url = 'https://www.theinfatuation.com/philadelphia/reviews'

infatuation_review_urls = []


#set up browser and parser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

browser.visit(infatuation_url)


for x in range(100):
    # retrieve philly restaurant urls
    time.sleep(2)
    browser.find_by_css('.styles_loadMoreButton__GN6yu').first.click()
    html = browser.html
    soup = bs(html, 'html.parser')
    
infatuation_review_urls.append([a['href'] for a in soup.find_all('a', class_="styles_absoluteLink__6MeI8", href=True)])
    

print('Finished')
```

```python
philly_reviews_df = pd.DataFrame(infatuation_review_urls)
philly_reviews_df.to_csv('philly_infatuation_review_urls.csv', index=False, encoding='utf-8')
```

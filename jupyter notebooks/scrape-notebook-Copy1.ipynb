```python
#scraping urls

infatuation_url = 'https://www.theinfatuation.com/philadelphia/reviews'

infatuation_review_urls = ''


#set up browser and parser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# set up collection

count = 0

browser.visit(infatuation_url)


for x in range(5):
    # retrieve michelin restaurant urls
    time.sleep(2)
    browser.find_by_css('.styles_button__r6aV0').first.click()
    html = browser.html
    soup = bs(html, 'html.parser')
    infatuation_review_urls.append([a['href'] for a in soup.find_all('a', class_="styles_absoluteLink__6MeI8", href=True)])
    count = count + 1
    print(count)
    

print('Finished')
```

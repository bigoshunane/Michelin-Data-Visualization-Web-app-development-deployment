```python
import pandas as pd
from currency_converter import CurrencyConverter

c = CurrencyConverter()
```

```python
# import csv

michelin = 'data/2021CountryContinent.csv'
michelin_df = pd.read_csv(michelin, encoding='ISO-8859-1')
michelin_df.head()
```

```python
# run currency converter function 

def currency_convertor(michelin_df,new_currency='USD'):
    try:
        
        # convert currency and store in new_curr variable
        
        price = michelin_df['MaxPrice']
        curr = michelin_df['Currency']
        new_curr = c.convert(price,curr,new_currency)
        return new_curr
    
    except:
        
        # log incompatible currencies
        
        print(f'{curr} failed')

#add column for max price in USD
michelin_df['max_usd'] = michelin_df.apply(lambda x: currency_convertor(x,new_currency='USD'), axis=1)
```

```python
# create bins for max_usd and add bin to new column

bins = [0, 300, 600, 900, 1000, 1500]
labels = [1,2,3,4,5]
michelin_df['price'] = pd.cut(michelin_df['max_usd'], bins=bins, labels=labels)

michelin_df
```
